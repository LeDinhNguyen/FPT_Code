## **CRAW FILE PROCESSING**

```python
crawdf.head(5)
```

```python
import re
dfc = pd.read_csv("craw.csv")
pattern = r"{'cast_id': ([^}]+), 'character': ([^}]+), 'credit_id': ([^}]+), 'gender': ([^}]+), 'id': ([^}]+), 'name': ([^}]+), 'order': ([^}]+), 'profile_path': ([^}]+)}"

cast_table = []
for row in range(len(dfc['cast'])):
    cast_id = re.findall(pattern=pattern,string=dfc['cast'][row])
    for i in cast_id:
        cast = {
            'cast_id': int(i[0]),
            'character': str(i[1][1:-1]).replace('"',""),
            'credit_id': str(i[2][1:-1]), 
            'gender': int(i[3]), 
            'id': int(i[4]), 
            'name': str(i[5][1:-1]).replace('"',""), 
            'order': int(i[6]), 
            'profile_path': str(i[7]).replace("'",""),
            'movie_id': dfc["id"][row]
        }
        cast_table.append(cast)
csttbl = pd.DataFrame(cast_table)

pattern2 = r"{'credit_id': ([^}]+), 'department': ([^}]+), 'gender': ([^}]+), 'id': ([^}]+), 'job': ([^}]+), 'name': ([^}]+), 'profile_path': ([^}]+)}"

crew_table = []
for row in range(len(dfc['crew'])):
    crew_id = re.findall(pattern=pattern2,string=dfc['crew'][row])
    for i in crew_id:
        crew = {
            'credit_id': str(i[0][1:-1]).replace('"',""),
            'department': str(i[1][1:-1]).replace('"',""),
            'gender': int(i[2]),
            'id': int(i[3]),
            'job': str(i[4][1:-1]).replace('"',""),
            'name': str(i[5][1:-1]).replace('"',""),
            'profile_path': str(i[6]).replace('"',"").replace("'",""),
            'movie_id': dfc["id"][row]
        }

        crew_table.append(crew)
crwtbl = pd.DataFrame(crew_table)
```

# **Movies Information file Processing**

```python
moviesinfodf['release_date'] = pd.to_datetime(moviesinfodf['release_date'], errors='coerce')
filtered_df = moviesinfodf.copy()
moviesinfodf.drop(columns=['belongs_to_collection', 'genres', 'production_companies',
       'production_countries', 'spoken_languages'], inplace=True)
moviesinfodf.head(5)
moviesinfodf['release_date'] = pd.to_datetime(moviesinfodf['release_date'], errors='coerce')
print(filtered_df.columns)
filtered_df.drop(columns=['release_date', 'revenue', 'runtime','status', 'tagline', 'title', 'video','vote_average', 'vote_count','adult','budget','homepage','imdb_id','original_language','original_title','overview','popularity','poster_path'], inplace=True)
filtered_df.columns
```

```python
patternCountry = r"{'iso_3166_1': ([^}]+), 'name': ([^}]+)}"
production_table = []
for row in range(len(filtered_df['production_countries'])):
    try:
        production_id = re.findall(pattern=patternCountry,string=filtered_df['production_countries'][row])
        for i in production_id:
            production = {
                'iso_3166_1': str(i[0][1:-1]).replace('"',""),
                'name': str(i[1][1:-1]).replace('"',""),
                'movie_id': filtered_df["id"][row]
            }
            production_table.append(production)
    except:
        pass
country_df = pd.DataFrame(production_table)
country_df.to_csv("../country.csv")
country_df
```

```python
filtered_df['genres'][0]

patternGenre = r"{'id': ([^{}]+), 'name': ([^{}]+)}"
genre_table = []

for row in range(len(filtered_df['genres'])):
    genre_id = re.findall(pattern=patternGenre,string=filtered_df['genres'][row])
    print(genre_id)
    for i in genre_id:
        genre = {
            'id': int(i[0]),
            'name': str(i[1][1:-1]).replace('"',""),
            'movie_id': filtered_df["id"][row]
        }
        genre_table.append(genre)

genre_df = pd.DataFrame(genre_table)
genre_df.to_csv("../genre.csv")
```

```python
genre_df.head(5)
```

```python
patternCollection = r"{'id': ([^{}]+), 'name': ([^{}]+), 'poster_path': ([^{}]+), 'backdrop_path': ([^{}]+)}"

collectiontbl = []
for row in range(len(filtered_df['belongs_to_collection'])):
    try:
        collection_id = re.findall(pattern=patternCollection,string=filtered_df['belongs_to_collection'][row])
        for i in collection_id:
            collection = {
                'id': int(i[0]),
                'name': str(i[1][1:-1]).replace("'",""),
                'poster_path': str(i[2][1:-1]).replace("'",""),
                'backdrop_path': str(i[3][1:-1]).replace("'",""),
                'movie_id': filtered_df['id'][row]
            }
            collectiontbl.append(collection)
    except:
        # print(row)
        pass

collection_df = pd.DataFrame(collectiontbl)
collection_df.to_csv("../collection_df.csv")
```

```python
collection_df.head(5)
```

```python
patternLanguage = r"{'iso_639_1': ([^{}]+), 'name': ([^{}]+)}"

languagetbl = []
for row in range(len(filtered_df['spoken_languages'])):
    try:
        language_id = re.findall(pattern=patternLanguage,string=filtered_df['spoken_languages'][row])
        for i in language_id:
            language = {
                'iso_639_1': str(i[0][1:-1]).replace("'",""),
                'name': str(i[1][1:-1]).replace("'",""),
                'movie_id': filtered_df['id'][row]
            }
            languagetbl.append(language)
    except:
        # print(row)
        pass

language_df = pd.DataFrame(languagetbl)
language_df.to_csv("../language_df.csv")
```
