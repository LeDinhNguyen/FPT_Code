# Preprocessing Data

```python
import pandas as pd
```

```python
df = pd.read_excel("./LAB DAN SO.xlsx",header=2)
df
```

```python
cols = df.columns
for i in range(2, len(cols), 2):
    df = df.rename(columns={cols[i+1]: cols[i]})
df
```

```python
cities = df["Unnamed: 0"]
cities.iloc[0] = "Thành phố"
```

```python
pd.concat([cities, df["2013"]], axis=1)
```

```python
data_by_year = {}
for i in range(1, len(cols), 3):
    year = cols[i]
    data_by_year[year] = pd.concat([cities, df[year]], axis=1)
    data_by_year[year].columns = data_by_year[year].iloc[0]
    data_by_year[year] = data_by_year[year][1:]
data_by_year["2011"]
```

```python
def add_year_col(data_by_year):
    for year in data_by_year.keys():
        df = data_by_year[year]
        years = [year for _ in range(df.shape[0])]
        df["Năm"] = years
        data_by_year[year] = df
    return data_by_year
data_by_year = add_year_col(data_by_year)
```

```python
final_df = pd.DataFrame()
for year in data_by_year.keys():
    final_df = pd.concat([final_df, data_by_year[year]], )
final_df["Năm"] = final_df["Năm"].str[:4]
final_df["Thành phố"].unique()
```

```python
final_df.to_csv("./final_data.csv", index=False)
```

### **Question 1:** Dùng biểu đồ line biểu diễn tỉ lệ tăng dân số trung bình của các năm của các tỉnh có tỉ lệ tăng nhanh nhất

```python
def grown_rate_by_area(df: pd.DataFrame, city: str) -> tuple:
    rates = np.array([0])
    years = df["Năm"].unique()
    for i in range(len(years)-1):
        start = years[i]
        end = years[i+1]
        start_population = df[(df["Thành phố"] == city) & (df['Năm'] == start)]
        end_population = df[(df["Thành phố"] == city) & (df['Năm'] == end)]
        rate = (end_population['Dân số trung bình (Nghìn người)'].values - start_population['Dân số trung bình (Nghìn người)'].values) / start_population['Dân số trung bình (Nghìn người)'].values * 100
        rates = np.concatenate((rates, rate), axis=0)
    rate_by_year = {
        "Tỉ lệ": rates,
        "Năm": years
    }

    return {city: pd.DataFrame(rate_by_year)}

grown_rate_by_area(df, "TP.Hồ Chí Minh")
```

```python
example_cities = ["TP.Hồ Chí Minh", "Hà Nội", "Đà Nẵng", "Quảng Nam", "Đồng Nai"]
for i in range(len(example_cities)):
    a = grown_rate_by_area(df, example_cities[i])[example_cities[i]]
    plt.plot(a["Năm"], a["Tỉ lệ"], label=f"{example_cities[i]}")
plt.legend()
plt.show()
```
