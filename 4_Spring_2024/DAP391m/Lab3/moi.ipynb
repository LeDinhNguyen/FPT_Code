```python
import pandas as pd

df = pd.read_excel("./LAB DAN SO.xlsx", header=2)
df
```

```python
df.columns
```

```python
cols = df.columns
for i in range(2, len(cols), 2):
    df = df.rename(columns={cols[i+1]: cols[i]})
```

```python
df = df.rename(columns={"Unnamed: 0": "Cities", "Unnamed: 1": "Code"})
```

```python
df["Cities"].iloc[0] = "Cities"
```

```python
df[["Cities", "Code", 2011]]
```

```python
# years = df["Năm"].unique()
areas = ['CẢ NƯỚC', 'Tây Nguyên', 'Đông Nam Bộ', 'Đồng bằng sông Hồng', 'Trung du và miền núi phía Bắc', 'Bắc Trung Bộ và Duyên hải miền Trung', 'Đồng bằng sông Cửu Long']
for area in areas:
    df.drop(df[df['Cities'] == area].index, inplace=True)
# df = df.iloc[1:]
```

```python
cols = df.columns
data_by_year = {}

for i in range(2, len(cols), 2):
    year = cols[i]
    data = df[["Cities", "Code", year]]
    data_by_year[year] = data
    data_by_year[year].columns = data_by_year[year].iloc[0]
    data_by_year[year] = data_by_year[year][1:]
    data_by_year[year]["Year"] = [year for _ in range(len(data)-1)]
data_by_year[2011]
```

```python
for year in data_by_year.keys():
    data = data_by_year[year]
    years = [year for _ in range(len(data))]
    data["Year"] = years
    data_by_year[year] = data
```

```python
final_data = pd.DataFrame()
for year in data_by_year.keys():
    final_data = pd.concat([final_data, data_by_year[year]])
final_data[final_data["Cities"] == "Hà Nội"]
```

```python
final_data = final_data.replace("2019 (*)", "2022")
final_data = final_data.replace("2020 (*)", "2022")
final_data = final_data.replace("2021 (*)", "2022")
final_data = final_data.replace("2022 (*)", "2022")
final_data
```
